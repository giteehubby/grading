```python
from typing import Union, Tuple, List
class CustomDataset(Dataset):
    def __init__(self, data:Union[Tuple,List]):
        self.data = data
 
    def __getitem__(self, index):
        # 获取单个样本，并进行数据预处理操作（如需要）
        sample = [data[index] for data in self.data]
        # 这里可以添加数据预处理代码，例如图像的缩放、裁剪、标准化等
        return sample
 
    def __len__(self):
        # 返回数据集的大小
        return len(self.data[0])
```

```python
def f1score(pred,label):

    with torch.no_grad():
        pred = torch.argmax(pred)
        TP = sum((label == 1) and (pred == 1))#真正
        FP = sum((label == 0) and (pred == 1))#假正
        FN = sum((label == 1) and (pred == 0))#假反
        precision = TP/(TP+FP)
        recall = TP/(TP+FN)
        return 2 * precision * recall/(precision + recall)
```
